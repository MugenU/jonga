## Example 3

Import modules and configure matplotlib output to notebook.

```python
import re
import io
import matplotlib.pyplot as plt
import matplotlib.image as mpim

import jonga

%matplotlib inline
```

Construct call tracer object with groups defined by the first part (i.e. before the first '.') of the fully qualified name of the function

```python
ct = jonga.CallTracer(dstflt='^(re.|sre_)', grpflt='^[^\.]*')
```

Start call tracing

```python
ct.start()
```

Call regex compile function

```python
rc = re.compile('^[^\.]*.[^\.]*')
```

Stop tracing

```python
ct.stop()
```

Define font family specification string

```python
fntfm = 'Vera Sans, DejaVu Sans, Liberation Sans, Arial, Helvetica, sans'
```

Get graph and construct PNG format visualisation image

```python
g = ct.graph(size='32,32', fntsz=12, fntfm=fntfm)

ct.graph('example3.svg', fntfm=fntfm)

png = g.draw(format='png')
bio = io.BytesIO()
bio.write(png)
bio.seek(0)
img = mpim.imread(bio)
```

Display graph

```python
fig = plt.figure(figsize=(16,16))
axim = plt.imshow(img)
ax = plt.axis('off')
plt.show()
```
